```python
# escribir y leer datos de un archivo

file1 =open("miarchivo.txt","w") 
L = ["Tango \n","Quebec \n","Mike \n"]  
# \n EOL (End of Line) 

file1.write("Hola \n") 
file1.writelines(L) 
file1.close()

#se cierra el archivo 
file1 =open("miarchivo.txt","r+")  

print("Salida de la funcion read ")
print(file1.read())
print

# seek(n) se posiciona en la enesima linea 
# en este caso el comienzo del archivo. 

file1.seek(0)  
print("Salida de leer una linea: ")
print(file1.readline())
print

file1.seek(0) 

# muestra la diferencia entre readline y read 

print("Salida de read(9): ")
print(file1.read(9))
print

file1.seek(0)

print("Salida de readline(9): ")
print(file1.readline(9)) 
file1.seek(0) 

# funcion readlinesprint("Salida de readlines: ")
print(file1.readlines()) 
print

file1.close()
```

```python
import matplotlib.pyplot as plt
import csv
x = []
y = []

with open('ejemplo.csv','r')as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y, label='Datos interesantes!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafica interesante\ncon dos picos')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

x, y = np.loadtxt('ejemplo.csv', delimiter=',', unpack=True)
plt.plot(x,y,label='Datos interesantes!')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafica interesante\ncon dos picos')
plt.legend()

# plt.savefig('picos.pdf') 
# si queremos guardar la grafica como un archivo PDF
# plt.savefig('picos.png') 
# si queremos guardar la grafica como un archivo de imagen

plt.show()
```
